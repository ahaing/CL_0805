```python
import numpy as np
import matplotlib.pyplot as plt 
figure = plt.figure

x = np.linspace(-3,3,50)
# linspace 線性的圖

display(x)
figure = plt.figure(figsize=(20,4))
axes = figure.add_subplot() # 甚麼意思??
axes.plot(x)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt 
figure = plt.figure

x = np.linspace(-5,5,50) # 產生 50 個
# linspace 線性的圖
x = x * 2 + 10
y = x ** 4

display(x)
figure = plt.figure(figsize=(20,4))
plotF = figure.add_subplot() # 甚麼意思??
plotF.plot(x)
plotF.plot(y)
plt.show()
```

```python
# 徑度

t = np.arange(0.,1.0, 0.05)
display(t)

y1 = np.sin(2 * np.pi * t) # 徑度換角度
display(y1)
y2 = np.cos(2 * np.pi * t) # 徑度換角度
display(y2)

figure1 = plt.figure(figsize=(8,4))
plotF = figure1.add_subplot()
# axes2 = figure1.add_subplot()
plotF.plot(y1)
plotF.plot(y2)
plt.show()
```
